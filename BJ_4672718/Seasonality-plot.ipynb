```python
import matplotlib.pyplot as plt
import pandas as pd

data=pd.read_csv("total_seasonality.csv")

data['year'] = pd.to_datetime(data['year'])


columns = [ 'N', 'T','F']
axes = data[columns].plot(subplots=True, title='Annual Seasonability Index plot')
for label in axes:
    label.set_ylabel('S.I')
    label.set_xlabel("year")
plt.savefig('seasonabilityPlot.jpg')
```
