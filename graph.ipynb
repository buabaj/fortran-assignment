```python
import matplotlib.pyplot as plt
import pandas as pd

graph=pd.read_csv("all_average.csv")
graph.head(4)




graph['month'] = pd.to_datetime(graph['month'])


cols_plot = [ 'north', 'trans','forest']
axes = graph[cols_plot].plot(marker='o', linestyle='-', figsize=(14, 8), subplots=True,color="red")
for label in axes:
    label.set_ylabel('Rainfall')
    label.set_xlabel("Month")
plt.savefig('time_series.jpg')

```
